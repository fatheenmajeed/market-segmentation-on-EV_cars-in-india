# hyperparameter tuning
#  finding 'K' value by Elbow Method

```python
from sklearn.cluster import KMeans
```

```python
inertia = []
range_val = range(1, 15)
for i in range_val:
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(pd.DataFrame(scalar_df))
    inertia.append(km.inertia_)

plt.plot(range_val, inertia,'bx-')  # Fixed variable name here
plt.xlabel('Values of K')  # Fixed typo in 'Values og K'
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```

```python
#so it falls on 4(at k value 4 the graph is taking a ELBOW)
```

# model building using KMeans

```python
kmeans=KMeans(4)
kmeans.fit(pd.DataFrame(scalar_df))
pca_df_kmeans=pd.concat([pca_df,pd.DataFrame({'cluster':kmeans.labels_})],axis=1)
pca_df_kmeans
```

```python
plt.figure(figsize=(8,8))
ax=sns.scatterplot(x='PCA1',y='PCA2',hue='cluster',data=pca_df_kmeans,palette='rainbow')
plt.title('clustering using kmeans algorithms')
plt.show()
```

```python
#find all cluster center
cluster_centers=pd.DataFrame(data=kmeans.cluster_centers_,columns=[df.columns])
#inverse transform the data
cluster_centers=scalar.inverse_transform(cluster_centers)
cluster_centers=pd.DataFrame(data=cluster_centers,columns=[df.columns])
cluster_centers
```

```python
#creating a target column
cluster_df=pd.concat([df,pd.DataFrame({'cluster':kmeans.labels_})],axis=1)
cluster_df
```
